# PDF embeddings and t-SNE (controller demo)

This notebook shows how to run `PdfEmbeddingAnalysis` inside Jupyter: install the Poetry environment, pick the kernel, run the pipeline, inspect arrays, and open Plotly figures.

**Prerequisite:** from the repo root, run `POETRY_KEYRING_ENABLED=false poetry install`.

**Kernel:** choose the project virtualenv (`.venv/bin/python`) or run `poetry run python -m ipykernel install --user --name=aweeumz` and pick the **aweeumz** kernel (see README).

```python
# Plotly output in Jupyter — pick one renderer for your setup
import plotly.io as pio

pio.renderers.default = "notebook_connected"  # try "browser" or "iframe_connected" if blank
```

```python
from pathlib import Path

from pdf_embedding_analysis import PdfEmbeddingAnalysis

# Folder that contains your PDFs (only top-level .pdf files are used)
PDF_DIR = Path("../data/Release_1").resolve()

# Cache directory for embeddings + t-SNE (delete files here to force recomputation)
EMBED_DIR = Path("../data/Release_1_embeddings").resolve()

PDF_DIR, EMBED_DIR
```

```python
analysis = PdfEmbeddingAnalysis(
    PDF_DIR,
    EMBED_DIR,
    debug=False,  # set True for verbose print diagnostics
)

analysis.run_analysis()
```

```python
import numpy as np

E = analysis.embeddings
T = analysis.tsne_3d
meta = analysis.metadata

print("embeddings:", E.shape, E.dtype)
print("t-SNE:", T.shape, T.dtype)
print("chunks:", len(meta))
print("example metadata keys:", list(meta[0].keys()))
print("first chunk preview:", meta[0]["text"][:120].replace("\n", " "), "…")
```

```python
# Legend entries = PDF filenames (one color series per file)
fig_by_file = analysis.plot_tsne_by_filename()
fig_by_file.show()
```

```python
# Alternative: single trace colored by chunk id, detailed hover
fig_chunks = analysis.plot_tsne_scatter_3d()
fig_chunks.show()
```

## Cache invalidation

- **New or edited PDFs:** remove `embeddings.npy` and `metadata.json` under `EMBED_DIR` (or delete the whole folder), then run `run_analysis()` again.
- **t-SNE only:** delete `tsne_3d.npy` to recompute the 3D layout from existing embeddings.
